```python
import requests
from bs4 import BeautifulSoup

# 날짜 및 pdf url 크롤링
total_info = []
for k in range(1, 90):
    # print(f'------------------------------------------------{k}페이지 크롤링------------------------------------------------')
    page = k
    URL = f'https://finance.naver.com/research/debenture_list.naver?keyword=&brokerCode=&searchType=writeDate&writeFromDate=2021-01-01&writeToDate=2023-12-31&x=40&y=21&page={page}'
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, 'html.parser')
    table_tags = soup.select('tr')
    info = []
    for table_tag in table_tags:
        try:
            title_tag = table_tag.select_one('td > a')
            pdf_tags = table_tag.select('td.file > a')
            date_tags = table_tag.select('td.date')
            if pdf_tags!=[] and date_tags!=[]:
                pdf = pdf_tags[0].attrs['href']
                date = date_tags[0].text
                title = title_tag.text
                pattern='\\d+\\.pdf'
                file_name = re.findall(pattern, pdf)
                file_name = date+'_'+file_name[0]
                info.append((pdf, date, title, file_name))
        except:
            title_tag = table_tag.select_one('td > a')
            print(f'error {title_tag}')
    total_info.extend(info)
total_info
```

```python
import pandas as pd

# 데이터프레임으로 변환
df = pd.DataFrame(total_info, columns=["pdf_link", "date", "title", "file_name"])
df
```

```python
# 데이터프레임을 CSV 파일로 저장
df.to_csv("pdf_link_crawl.csv", sep='\t', index=False)  # index=False를 사용하여 인덱스를 저장하지 않음

print("CSV 파일이 저장되었습니다.")
```

# pdf 파일의 내용을 csv 파일에 저장

```python
import pandas as pd

# CSV 파일을 불러와서 데이터프레임으로 변환
df = pd.read_csv("pdf_link_crawl.csv", sep='\t')

# 데이터프레임을 리스트 안의 튜플 데이터 형태로 변환
total_info = [tuple(row) for row in df.values]
total_info
```

```python
df['content_file'] = df['file_name'].apply(lambda x : x.replace('.pdf', '.txt'))
df
```

```python
# 데이터프레임을 CSV 파일로 저장
df.to_csv("pdf_link_crawl_add_txt.csv", sep='\t', index=False)  # index=False를 사용하여 인덱스를 저장하지 않음

print("CSV 파일이 저장되었습니다.")
```

```python
import fitz
dir = './reportpdf/'
content = ''
for i in range(len(df)):
    df_info_list = df.iloc[i].to_list()
    pdf_name = dir + df_info_list[3]
    with fitz.open(pdf_name) as doc:
        for page in doc:
            content += page.get_text()
            txt_name = './reporttxt/' + df_info_list[4]
            with open(txt_name, 'w') as f:
                f.write(content)
```
